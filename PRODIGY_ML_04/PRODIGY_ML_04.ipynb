```python
import os
import zipfile
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# STEP 1: LOAD DATA
# ---------------------------------------------------------
# We use the 'Rock-Paper-Scissors' dataset (a standard Hand Gesture dataset)
# because the original LeapGestRecog is too large (2GB+) for free Colab sessions.

print("Downloading dataset...")
# Download Training Data
# Corrected URL for rps.zip
!wget --no-check-certificate \
    https://storage.googleapis.com/download.tensorflow.org/data/rps.zip \
    -O /tmp/rps.zip

# Download Test Data
# Corrected URL for rps-test-set.zip
!wget --no-check-certificate \
    https://storage.googleapis.com/download.tensorflow.org/data/rps-test-set.zip \
    -O /tmp/rps-test-set.zip

# Unzip
local_zip = '/tmp/rps.zip'
zip_ref = zipfile.ZipFile(local_zip, 'r')
zip_ref.extractall('/tmp/')
zip_ref.close()

local_zip = '/tmp/rps-test-set.zip'
zip_ref = zipfile.ZipFile(local_zip, 'r')
zip_ref.extractall('/tmp/')
zip_ref.close()

print("Dataset downloaded and extracted!")

train_dir = '/tmp/rps/'
test_dir = '/tmp/rps-test-set/'

# ---------------------------------------------------------
# STEP 2: PREPROCESSING (Data Augmentation)
# ---------------------------------------------------------
# We rescale images to 0-1 range and add rotation/shifts to make the model smarter
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

validation_datagen = ImageDataGenerator(rescale=1./255)

# Load images from folders
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical' # 3 classes: Rock, Paper, Scissors
)

validation_generator = validation_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)

# ---------------------------------------------------------
# STEP 3: BUILD THE CNN MODEL
# ---------------------------------------------------------
# This is a Deep Learning model (Convolutional Neural Network)
model = tf.keras.models.Sequential([
    # First Convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),

    # Second Convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    # Third Convolution
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    # Flatten the results to feed into a DNN
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5), # Avoid overfitting

    # Hidden Layer
    tf.keras.layers.Dense(512, activation='relu'),

    # Output Layer (3 units for Rock, Paper, Scissors)
    tf.keras.layers.Dense(3, activation='softmax')
])

model.summary()

model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

# ---------------------------------------------------------
# STEP 4: TRAIN THE MODEL
# ---------------------------------------------------------
print("\nStarting Training... (This will take about 2-3 minutes)")
history = model.fit(
    train_generator,
    epochs=10,  # 10 loops over the data
    validation_data=validation_generator,
    verbose=1
)

print("Training Complete!")

# ---------------------------------------------------------
# STEP 5: VISUALIZE RESULTS
# ---------------------------------------------------------
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.figure(figsize=(12, 6))

# Plot Accuracy
plt.subplot(1, 2, 1)
plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and Validation Accuracy')
plt.legend(loc=0)

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(epochs, loss, 'r', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.legend(loc=0)

plt.show()

# ---------------------------------------------------------
# STEP 6: TEST PREDICTION
# ---------------------------------------------------------
# Let's predict a single image from the test set to prove it works
import numpy as np
from tensorflow.keras.preprocessing import image

# Load a sample image (Scissors)
sample_path = '/tmp/rps-test-set/scissors/testscissors01-04.png'
img = image.load_img(sample_path, target_size=(150, 150))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)

images = np.vstack([x])
classes = model.predict(images, batch_size=10)

print("\nPrediction Result:")
print(classes)
print("(Order: Paper, Rock, Scissors)")
```
